```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Datasets/bank_transactions_data_2.csv", encoding="utf-8")
```

## Anomoly Discorver

```python
# Occupation-Based Spending Distribution

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="CustomerOccupation", y="TransactionAmount")
plt.title("Occupation-Based Spending Distribution")
plt.show()
```

```python
# Age group / transaction amount

def ageGroup(age):
    if age < 20:
        return "teen"
    elif 20 <= age < 40:
        return "working_age"
    elif 40 <= age <= 60:
        return "middle_age"
    elif age > 60:
        return "senior"
    
df["Age_Group"] = df["CustomerAge"].apply(ageGroup)
df.head()

sns.boxplot(data=df, x="Age_Group", y="TransactionAmount")
plt.title("Boxplot of transaction amount per age group")
plt.show()
```

```python
# Transactions per type
sns.countplot(data=df, x="TransactionType")
plt.title("Count of no of transaction types used")
plt.show()
```

```python
# Login Attemps for Age Group
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x="Age_Group", hue="LoginAttempts")
plt.title("Count of login attempts for age group")
plt.show()
```

```python
# DeviceID and IPAddresses per account id
# 495 acc id with more than 2 device id
device_counts = df.groupby('AccountID')['DeviceID'].nunique().reset_index()
print(device_counts)
ip_add_counts = df.groupby('AccountID')['IP Address'].nunique().reset_index()
print(ip_add_counts)
```
